```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
print("LIB imported")
```

```python
data1 = pd.read_csv("C:\\Users\\LENOVO\\Downloads\\data.csv")
data1.tail()
```

```python
Data_X = data1[["Rainfall","Temperature","Ph"]]
Data_X.head()
```

```python
Data_Y = pd.DataFrame(data1["Production"])
# Data_Y = data1["Production"].sample(frac=1)
Data_Y.head()
```

Starting the plotting

```python
data1.describe()
```

```python
plt.scatter(Data_X["Rainfall"],Data_Y)
```

```python
lm_model = LinearRegression().fit(Data_X,Data_Y)
```

```python
predict = lm_model.predict(Data_X)
predict_frame = pd.DataFrame(predict)
type(predict_frame)
```

```python
plt.plot(predict,c='red')
plt.xlabel("Index")
plt.ylabel("Predict")
plt.grid(True)
```
